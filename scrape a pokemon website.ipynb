```python
import requests
```

```python
url = "https://pokemondb.net/pokedex/all"
```

```python
result = requests.get(url)
print(result)
```

```python
data = result.content
print(data)
```

```python
from bs4 import BeautifulSoup as soup
```

```python
data = soup(data, 'html.parser')
print(data)
```

```python
data.h1
```

```python
table = data.findAll('table', {})
print(len(table))
```

```python
table = table[0]
```

```python
headers = table.findAll('th', {})
print(len(headers))
print(headers)
```

```python
column_titles = [ct.text for ct in headers]
print(column_titles)
```

```python
rows_data = table.findAll('tr', {})[1:]
print(len(rows_data))
print(rows_data)
```

```python
first_row = rows_data[0].findAll('td', {})
for d in first_row:
    print(d.text)
```

```python
table_rows = []
for row in rows_data:
    current_row = []
    d = row.findAll('td', {})
    for i in d:
        current_row.append(i.text)
    table_rows.append(current_row)
```

```python
print(table_rows)
```

```python
with open("pokemon.csv", 'w', encoding = 'UTF-8') as file:
    header_string = ','.join(column_titles)
    header_string += "\n"
    file.write(header_string)
    
    for row in table_rows:
        row_string = ','.join(row)
        row_string += "\n"
        file.write(row_string)
```
