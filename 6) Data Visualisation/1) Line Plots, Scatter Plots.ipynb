```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plots

```python
x = np.arange(10)
print(x)

y1 = x**2
print(y1)

y2 = 2*x + 3
print(y2)
```

```python
plt.plot(x, y1)
plt.show()
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(x, y2, color = 'green')
plt.show()
```

```python
plt.plot(x, y1, color = 'blue', label = "Mango", marker = 'o')
plt.plot(x, y2, color = 'green', label = "Banana", linestyle = "dashed", marker = '^')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.legend()
plt.show()
```

```python
plt.plot(x, y1, color = 'blue')
plt.show()

plt.plot(x, y2, color = 'green')
plt.show()
```

```python
prices = np.array([1, 2, 3, 4])**2
print(prices)
```

```python
plt.plot(prices, marker = 'o')
plt.show()
```

## Scatter Plots

```python
plt.figure(figsize = (8, 8))  # Adjusting size of the plot
plt.scatter(x, y1, color = "red", label = "Apple", marker = "^")
plt.scatter(x, y2, color = "green", label = "Guava", marker = "o")
plt.xlabel("Price")
plt.ylabel("Time")
plt.title("Prices of Fruits over Time")
plt.legend()
plt.show()
```
