```python
import numpy as np
import matplotlib.pyplot as plt
```

## Univariate Normal Distribution

```python
# Generate a Normal Distribution (1-D)

u = 5       # mean
sigma = 1   # standard deviation

vals = u + sigma*np.random.randn(200)
print(vals.shape)
# print(vals)
```

```python
plt.hist(vals, color = "orange", edgecolor = "black")
plt.xlabel("X")
plt.ylabel("F(X)")
plt.title("Normal Distribution")
plt.show()
```

```python
x = vals
y = np.zeros(x.shape)

plt.scatter(x, y)
plt.show()
```

## Multivariate Normal Distribution

```python
# Bivariate Normal Distribution

mean = np.array([0, 0])
cov = np.array([[1, 0], [0, 1]])

dist = np.random.multivariate_normal(mean, cov, 500)
print(dist.shape)
# print(dist)
```

```python
plt.scatter(dist[ : , 0], dist[ : , 1])
plt.show()
```

```python
mean = np.array([0, 0])
cov = np.array([[1, 0.5], [0.5, 1]])

mean2 = np.array([5, 6])
cov2 = np.array([[1.3, 0.2], [0.2, 1.1]])

dist = np.random.multivariate_normal(mean, cov, 500)
dist2 = np.random.multivariate_normal(mean2, cov2, 500)
```

```python
plt.scatter(dist[ : , 0], dist[ : , 1])
plt.scatter(dist2[ : , 0], dist2[ : , 1])
plt.show()
```
