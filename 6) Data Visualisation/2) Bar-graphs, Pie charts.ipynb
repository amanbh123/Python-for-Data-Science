```python
import matplotlib.pyplot as plt
import numpy as np
```

## Bar-graphs

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
population = [100, 120, 180, 250, 300, 370]
```

```python
plt.bar(year, population, width = 0.5, color = 'green')
plt.xlabel("Year")
plt.ylabel("Popupation")
plt.title("Variation of Population over Years")
plt.show()
```

## Pie Charts

```python
subjects = ["Maths", "Chem", "Physics", "English"]
weightage = [20, 10, 15, 5]
colors = ["green", "orange", "blue", "yellow"]
explode = (0, 0.4, 0, 0)
plt.pie(weightage, labels = subjects, colors = colors, explode = explode, autopct = '%.2f%%', counterclock = False)
plt.title("Weightages of subjects in exam")
plt.show()
```
