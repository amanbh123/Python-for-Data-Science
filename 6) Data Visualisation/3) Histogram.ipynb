```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating a Standard Normal Distribution
Xsn = np.random.randn(100)
print(Xsn)
```

```python
# Creating a normal distribution from the standard normal distribution
sigma = 5  # Standard deviation
u = 60     # Mean
X = np.round(Xsn*sigma + u)
print(X)
```

```python
plt.style.use('seaborn')
```

```python
plt.hist(X)
plt.xlabel("Marks Range")
plt.ylabel("Prob. of Students")
plt.title("Demo Histogram")
plt.show()
```

```python
plt.hist(X, edgecolor = 'black', color = 'orange')

a = np.arange(50, 78, 2.7)
plt.xticks(a)

plt.xlabel("Marks Range")
plt.ylabel("Prob. of Students")
plt.title("Demo Histogram")
plt.show()
```

```python
bins = np.arange(50, 76, 5)
plt.hist(X, bins = bins, edgecolor = 'black', color = 'green')
plt.xlabel("Marks Range")
plt.ylabel("Prob. of Students")
plt.title("Demo Histogram")
plt.show()
```
