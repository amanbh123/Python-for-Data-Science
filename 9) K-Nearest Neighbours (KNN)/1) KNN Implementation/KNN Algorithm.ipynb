```python
X = dfx.values
Y = dfy.values
X = X[:,1:]
Y = Y[:,1]
print(X.shape)
print(Y.shape)
```

```python
query = np.array([2,3])
plt.scatter(query[0], query[1], color = "red")
plt.scatter(X[:,0], X[:,1], c = Y)
plt.show()
```

```python
def dist(x1, x2):
    return np.sqrt(sum((x1 - x2)**2))
    
def knn(X, Y, queryPoint, k = 5):
    vals = []
    m = X.shape[0]
    
    for i in range(m):
        d = dist(queryPoint, X[i])
        vals.append([d, Y[i]])
        
    vals.sort()
    vals = np.array(vals[:k])   # K-nearest points
    
    new_vals = np.unique(vals[:,1], return_counts = True)
    index = new_vals[1].argmax()
    pred = int(new_vals[0][index])
    
    return pred
```

```python
knn(X, Y, query)
```

```python
knn(X, Y, [0,4])
```

```python
knn(X, Y, [1,1])
```

```python
knn(X, Y, [2,1])
```
