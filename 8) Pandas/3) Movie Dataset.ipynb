## Working on [Movie dataset](https://www.kaggle.com/karrrimba/movie-metadatacsv)
Task : Mark a scatter plot b/w length of movie title (X-axis) and frequency of occurence of that length (Y-axis)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movie_metadata.csv")
```

```python
df.head(n=10)
```

```python
df.columns
```

```python
titles = list(df.get('movie_title'))
print(type(titles))
# print(titles)
```

```python
freq = {}
for title in titles:
    length = len(title)
    if freq.get(length) is None:
        freq[length] = 1
    else:
        freq[length] += 1
print(freq)
```

```python
X = np.array(list(freq.keys()))
Y = np.array(list(freq.values()))
```

```python
plt.scatter(X, Y, color = "green")
plt.xlabel("Lengths")
plt.ylabel("Freq")
plt.title("Movie Data Visualisation")
plt.show()
```
