```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## 1. Bar Plot

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std)
plt.show()
```

## 2. Count Plot

```python
sns.countplot(x = 'sex', data = tips)
plt.show()
```

```python
sns.countplot(x = 'time', data = tips)
plt.show()
```

## 3. Box Plot

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
plt.show()
```

```python
sns.boxplot(x = 'sex', y = 'total_bill', data = tips, hue = 'smoker')
plt.show()
```

## 4. Violin Plot

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
plt.show()
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker', split = True)
plt.show()
```

## 5. Distribution Plots

### (i) distplot

```python
sns.distplot(tips['total_bill'])
plt.show()
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 15, color = 'green')
plt.show()
```

### (ii) kdeplot

```python
sns.kdeplot(tips['total_bill'], color = 'orange')
plt.show()
```

### (iii) jointplot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips) # here, kind is scatter
plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex', color = 'green')
plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg', color = 'orange')
plt.show()
```

### (iv) Pair Plot

```python
sns.pairplot(tips, hue = 'sex')
```

## 6. Heat maps

```python
tips.head()
```

```python
tips_corr = tips.corr()
```

```python
tips_corr
```

```python
sns.heatmap(tips_corr)
plt.show()
```

```python
sns.heatmap(tips_corr, annot = True)
plt.show()
```

## Pivot Table

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
flights_pivot = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers') 
```

```python
flights_pivot
```

```python
sns.heatmap(flights_pivot, cmap = 'coolwarm')
plt.show()
```
