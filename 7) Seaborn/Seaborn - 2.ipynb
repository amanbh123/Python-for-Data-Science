```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

## 1. Bar Plot

```python
sns.barplot(x = 'sex', y = 'age', data = titanic)
plt.show()
```

```python
sns.barplot(x = 'sex', y = 'age', data  = titanic, estimator = np.std)
plt.show()
```

## 2. Count Plot

```python
sns.countplot(x = 'alive', data = titanic)
plt.show()
```

## 3. Box Plot

```python
sns.boxplot(x = 'sex', y = 'age', data = titanic)
plt.show()
```

```python
sns.boxplot(x = 'sex', y = 'age', data = titanic, hue = 'alive')
plt.show()
```

## 4. Violin Plot

```python
sns.violinplot(x = 'sex', y = 'age', data = titanic, hue = 'alone')
plt.show()
```

## 5. distplot

```python
sns.distplot(titanic['age'], color = 'orange')
plt.show()
```

```python
sns.distplot(titanic['age'], color = 'red', kde = False, bins = 10)
plt.show()
```

## 6. kdeplot

```python
sns.kdeplot(titanic['age'], color = 'green')
plt.show()
```

## 7. Joint Plot

```python
sns.jointplot(x = 'age', y = 'fare', data = titanic, kind = 'reg')
plt.show()
```

## 8. Heat Maps

```python
titanic_corr = titanic.corr()
```

```python
titanic_corr
```

```python
sns.heatmap(titanic_corr, annot = True)
plt.show()
```

## Pivot Table

```python
titanic_pivot = titanic.pivot_table(index = 'sex', columns = 'class', values = 'fare')
```
